```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu, sigma = 0.0, 0.1

fig = plt.figure(figsize = (12, 3))
ax = plt.subplot(1, 3, 1)
s = np.random.normal(mu, sigma, 500)
count, bins, ignored = plt.hist(s, 30, density = True, color = 'k', alpha = 0.3)
ax.plot(bins[:-1:] + np.diff(bins) / 2.0, count, 'c-')
ax.plot(bins, 1.0 / (sigma * np.sqrt(2.0 * np.pi)) *
               np.exp( - (bins - mu) ** 2 / (2.0 * sigma ** 2) ), 'r--')

ax.grid()
ax.set_xlabel(r"$X_i$")
ax.set_ylabel(r"frequency")
ax.set_xlim(-0.4, 0.4)
ax.set_ylim(0, 4.5)
ax.set_title(r"$N = 500$")

ax = plt.subplot(1, 3, 2)
s = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 30, density = True, color = 'k', alpha = 0.3)
plt.plot(bins[:-1:] + np.diff(bins) / 2.0, count, 'c-')
plt.plot(bins, 1.0 / (sigma * np.sqrt(2.0 * np.pi)) *
               np.exp( - (bins - mu) ** 2 / (2.0 * sigma ** 2) ), 'r--')

plt.grid()
ax.set_xlabel(r"$X_i$")
ax.set_xlim(-0.4, 0.4)
ax.set_ylim(0, 4.5)
ax.set_title(r"$N = 1000$")

ax = plt.subplot(1, 3, 3)
s = np.random.normal(mu, sigma, 4000)
count, bins, ignored = plt.hist(s, 30, density = True, color = 'k', alpha = 0.3)
plt.plot(bins[:-1:] + np.diff(bins) / 2.0, count, 'c-')
plt.plot(bins, 1.0 / (sigma * np.sqrt(2.0 * np.pi)) *
               np.exp( - (bins - mu) ** 2 / (2.0 * sigma ** 2) ), 'r--')

plt.grid()
ax.set_xlabel(r"$X_i$")
ax.set_xlim(-0.4, 0.4)
ax.set_ylim(0, 4.5)
ax.set_title(r"$N = 4000$")

fig.savefig("../figures/gaussian_bar_graph.png", dpi = 150)
```
